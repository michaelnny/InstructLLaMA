```python
import numpy as np
import pickle

%matplotlib widget

from matplotlib import pyplot as plt
```

```python
def count_token_length(dataset_file):
    dataset = pickle.load(open(dataset_file, "rb"))
    token_length = []

    for pair in dataset:
        token_length.append(len(pair["prompt_tokens"]) + len(pair["completion_tokens"]))

    return token_length


def plot(token_length, title):
    # Plotting the token length distribution
    fit = plt.figure()
    plt.hist(token_length, bins=100, edgecolor='black')  # Adjust the number of bins as per your preference
    plt.xlabel('Token Length')
    plt.ylabel('Frequency')
    plt.title(title)
```

```python
def plot_token_length_distribution(dataset_file, name=""):

    token_length = count_token_length(dataset_file)

    print(f'mean: {np.mean(token_length)}, std: {np.std(token_length)}')
    print(f'min: {np.min(token_length)}, max: {np.max(token_length)}')

    plot(token_length, f'{name} - Token Length Distribution')

def plot_combined_token_length_distribution(dataset_files):
    token_length = []
    for dataset_file in dataset_files:
        token_length.extend(count_token_length(dataset_file))

    print(f'mean: {np.mean(token_length)}, std: {np.std(token_length)}')
    print(f'min: {np.min(token_length)}, max: {np.max(token_length)}')

    plot(token_length, f'Combined {len(dataset_files)} - Token Length Distribution')
```
